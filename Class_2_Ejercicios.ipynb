Por lo tanto, la probabilidad de que los amigos se encuentren es del 43.75%.

```python
import numpy as np
import matplotlib.pyplot as plt

# Par√°metros
espera = 15  # minutos que espera el primero que llega

# Generar puntos aleatorios para las llegadas de los dos amigos
np.random.seed(42)
n_puntos = 100000
x = np.random.uniform(0, 60, n_puntos)
y = np.random.uniform(0, 60, n_puntos)

# Calcular la diferencia absoluta entre los tiempos de llegada
diferencia = np.abs(x - y)

# Determinar cu√°ntos puntos cumplen con la condici√≥n de encontrarse
encuentros = diferencia <= espera
probabilidad = np.mean(encuentros)

# Visualizaci√≥n
plt.figure(figsize=(6, 6))
plt.scatter(x[encuentros], y[encuentros], color='green', s=1, label='Encuentros')
plt.scatter(x[~encuentros], y[~encuentros], color='red', s=1, label='No encuentros')
plt.title(f'Probabilidad de encuentro: {probabilidad:.4f}')
plt.xlabel('Tiempo de llegada del primer amigo (minutos)')
plt.ylabel('Tiempo de llegada del segundo amigo (minutos)')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir la probabilidad
print(f'Probabilidad de que los amigos se encuentren: {probabilidad:.4f}')
```
